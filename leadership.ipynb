```python
def remove_outlayer(col_name):
    Q1=np.percentile(df[col_name],25,method='midpoint')
    Q3=np.percentile(df[col_name],75,method='midpoint')
    IQR=Q3-Q1
    upper=Q3+1.5*IQR
    lower=Q1-1.5*IQR
    for i in df[col_name].index:
        if df.loc[i,col_name] >= upper:
            df.loc[i,col_name] = df[col_name].mean()
    for i in df[col_name].index:
        if df.loc[i,col_name] <= lower:
            df.loc[i,col_name] = df[col_name].mean()
            
```

```python
remove_outlayer('TeamSize')
remove_outlayer('Communication')
remove_outlayer('LearningCulture')
remove_outlayer('RewardFairness')
remove_outlayer('DecisionSpeed')
remove_outlayer('Stress')
remove_outlayer('Burnout')
remove_outlayer('TurnoverIntent')
```

```python
from sklearn.preprocessing import StandardScaler
scale_cluster=StandardScaler()
scale_cluster.fit(x)
x=scale_cluster.fit_transform(x)
x
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
wcss=[]
silhouette=[]
k=range(2,10)
for i in k:
    km=KMeans(n_clusters=i,init='k-means++',random_state=42)
    km.fit(x)
    wcss.append(km.inertia_)
    silhouette.append(silhouette_score(x,km.labels_))
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12,5))

ax[0].plot(k, wcss, marker='o')
ax[0].set_title('Elbow Method (Inertia)')
ax[0].set_xlabel('Number of Clusters k')
ax[0].set_ylabel('Inertia')



ax[1].plot(k, silhouette, marker='o')
ax[1].set_title('Silhouette Score')
ax[1].set_xlabel('Number of Clusters k')
ax[1].set_ylabel('Score')


plt.tight_layout()
plt.show()
```

```python
km=KMeans(n_clusters=4,init='k-means++',random_state=42)
clusters=km.fit_predict(x)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
from sklearn.naive_bayes import GaussianNB
nav=GaussianNB()
from sklearn.svm import SVC
sv=SVC()
from sklearn.ensemble import RandomForestClassifier
rf_c=RandomForestClassifier(criterion='entropy')
from sklearn.tree import DecisionTreeClassifier
d_tree_c=DecisionTreeClassifier(criterion='entropy')
models=[knn,nav,sv,rf_c,d_tree_c]
for model in models:
    print(model)
    model.fit(x_c_train,y_c_train)
    y_c_pred=model.predict(x_c_test)
    print(confusion_matrix(y_c_test,y_c_pred))
    print(accuracy_score(y_c_test,y_c_pred))
    print(classification_report(y_c_test,y_c_pred))
```
